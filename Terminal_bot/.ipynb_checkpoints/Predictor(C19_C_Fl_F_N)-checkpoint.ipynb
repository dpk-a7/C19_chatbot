```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_excel('COVID_DS1.xlsx')
X = df.iloc[:,:14].values
y = df.iloc[:,14:].values
```

```python
print(y[0][1])
```

```python
def make_it_one(y_,c,z):
    for i in range(0,len(y_)):
        if y_[i][c] == 1:
            y_[i][c] = z
        else:
            pass
    return y_
```

```python
j = make_it_one(y,0,1)
print(j)
```

```python
j = make_it_one(y,1,2)
print(j)
```

```python
j = make_it_one(y,2,3)
print(j)
```

```python
j = make_it_one(y,3,4)
print(j)
```

```python
print(j[0])
def map_it(j_):
    y_re = []
    for i in range(0,len(j_)):
        y_r = j_[i][0] + j_[i][1] + j_[i][2] + j_[i][3]
        y_re.append(y_r)
    return y_re
```

```python
#print(map_it(j))
y_ready  = map_it(j)
y_ready = np.array(y_ready)
print(len(y_ready))
```

```python
X_train,X_val,y_train,y_val = train_test_split(X,y_ready,test_size = 0.25)

classifier = CatBoostClassifier()
classifier.fit(X_train,y_train)
```

```python
y_pred  = classifier.predict(X_val)

print(np.concatenate((y_pred.reshape(len(y_pred),1),y_val.reshape(len(y_val),1)),1))
```

```python
cm = confusion_matrix(y_val,y_pred)
print(cm)
accuracy_score(y_val,y_pred)
```

```python
accuracy = cross_val_score(estimator = classifier,X = X_train,y = y_train,cv = 5)
print("Accuracy : ",accuracy.mean()*100)
print("Standard Deviation : ",accuracy.std()*100)
```
